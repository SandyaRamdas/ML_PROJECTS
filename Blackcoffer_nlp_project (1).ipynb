```python
## use requests and beautifulsoup to scrape to scrape the web pages
import os
import requests
from bs4 import BeautifulSoup
```

```python
## Create a folder to store text files
if not os.path.exists('article_text'):
  os.makedirs('article_text')
```

```python
## Function to exctract title and article text
def extract_article(url):
  try:
    response = requests.get(url) #requesting the webpage
    soup = BeautifulSoup(response.text, 'html.parser') # parsing the html content

    # Exctract the title
    title = soup.find('h1').get_text(strip=True)

    # Exctract article text
    article = " ".join([p.get_text(strip=True) for p in soup.find_all('p')])

    return title, article

  except Exception as e:
    print(f"Error exctracting {url}: {e}")
    return None, None
```

```python
# Iterate over each url
for index, row in df.iterrows():
  url_id = row['URL_ID']
  url = row['URL']

  title, article = extract_article(url)

  #save to a text file
  if title and article:
    file_path = os.path.join('article_text', f'{url_id}.txt')
    with open(file_path, 'w', encoding='utf-8') as file:
      file.write(f"{title}\n\n{article}")
      print(f'Article saved {url_id}.txt')
  else:
    print(f'Failed to extract data for url_id:{url_id}')
```

```python
#path to the zipped stopwords file
stopwords_zip_path = '/content/drive/MyDrive/StopWords-20241210T094826Z-001.zip'
```

```python
## Function to load stopwords from all files in the zipped folder
def load_stopwords_from_zip(zip_path):
  stopwords = set()
  with zipfile.ZipFile(stopwords_zip_path, 'r') as zip_file:
    #iterate over each file in the zip
    for file_name in zip_file.namelist():
      if file_name.endswith('.txt'):
        with zip_file.open(file_name) as file:
          #decode and process lines from the file
          file_stopwords = {line.decode('utf-8', errors='replace').strip().lower() for line in file if line.strip()}
          stopwords.update(file_stopwords)
  return stopwords
```

```python
##load all stopwords
custom_stopwords = load_stopwords_from_zip(stopwords_zip_path)
```

```python
##Function to clean text by removing stopwords
def clean_text(text, stopwords):
  text = text.lower() # convert to lowercase
  text = re.sub(r'[^\w\s]', '', text) # removes punctuation
  words = text.split() #tokenize the text
  filtered_words = [word for word in words if word not in stopwords] #removes stopwords
  cleaned_text = ' '.join(filtered_words)
  return cleaned_text
```

```python
#process each file in the cleaned_text_dir
for file_name in os.listdir(cleaned_text_dir):
  file_path = os.path.join(cleaned_text_dir, file_name)
  # read the cleaned file content
  with open(file_path, 'r') as f:
    text = f.read().lower()
    words = text.split() #tokenize the text
  #initialize counters
  positive_score = 0
  negative_score = 0

  # calculate positive and negative scores
  for word in words:
    if word in positive_words_dict:
      positive_score +=1
    elif word in negative_words_dict:
      negative_score +=1

  #calculate derived variables
  polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
  subjectivity_score = (positive_score + negative_score) / (len(words) + 0.000001)

  #store results
  derived_variables[file_name] = {
      'positive_score': positive_score,
      'negative_score': negative_score,
      'polarity_score': polarity_score,
      'subjectivity_score': subjectivity_score
  }

#print the results
for file_name, results in derived_variables.items():
  print(f"File: {file_name}")
  print(f"Positive Score: {results['positive_score']}")
  print(f"Negative Score: {results['negative_score']}")
  print(f"Polarity Score: {results['polarity_score']}")
  print(f"Subjectivity Score: {results['subjectivity_score']}")
  print("-" * 50)

```

```python
import os
import re
import csv
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')  # Ensure necessary resources are available

# Path to the directory containing cleaned text files
cleaned_text_dir = "/content/cleaned_text_dir"  # Replace with your actual path
output_csv = "analysis_results.csv"  # Output CSV file name

# Helper functions
def is_complex_word(word):
    vowels = "aeiou"
    word = word.lower()
    syllable_count = 0
    for i in range(len(word)):
        if word[i] in vowels and (i == 0 or word[i - 1] not in vowels):
            syllable_count += 1
    if word.endswith(("es", "ed")) and syllable_count > 1:
        syllable_count -= 1
    return syllable_count > 2

def count_syllables(word):
    vowels = "aeiou"
    word = word.lower()
    syllable_count = 0
    for i in range(len(word)):
        if word[i] in vowels and (i == 0 or word[i - 1] not in vowels):
            syllable_count += 1
    if word.endswith(("es", "ed")) and syllable_count > 1:
        syllable_count -= 1
    return syllable_count

def count_personal_pronouns(text):
    personal_pronouns = ["i", "we", "my", "ours", "us"]
    words = word_tokenize(text.lower())
    return sum(1 for word in words if word in personal_pronouns)

# Initialize results list
results = []

# Process each file in cleaned_text_dir
for file_name in os.listdir(cleaned_text_dir):
    file_path = os.path.join(cleaned_text_dir, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.txt'):
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                text = f.read().strip()

            sentences = sent_tokenize(text)
            words = word_tokenize(text)

            total_words = len(words)
            total_sentences = len(sentences)
            complex_word_count = sum(1 for word in words if is_complex_word(word))
            syllables_per_word = sum(count_syllables(word) for word in words) / total_words if total_words else 0
            avg_word_length = sum(len(word) for word in words) / total_words if total_words else 0
            personal_pronouns = count_personal_pronouns(text)

            avg_sentence_length = total_words / total_sentences if total_sentences else 0
            pct_complex_words = complex_word_count / total_words if total_words else 0
            fog_index = 0.4 * (avg_sentence_length + pct_complex_words)

            # Dummy scores for positive, negative, polarity, and subjectivity (replace with actual calculations if needed)
            positive_score = sum(1 for word in words if word in positive_words_dict)
            negative_score = sum(1 for word in words if word in negative_words_dict)
            polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) or 1)
            subjectivity_score = (positive_score + negative_score) / (total_words or 1)

            results.append({
                "File Name": file_name,
                "Positive Score": positive_score,
                "Negative Score": negative_score,
                "Polarity Score": polarity_score,
                "Subjectivity Score": subjectivity_score,
                "Avg Sentence Length": avg_sentence_length,
                "Percentage of Complex Words": pct_complex_words,
                "Fog Index": fog_index,
                "Avg Words Per Sentence": avg_sentence_length,
                "Complex Word Count": complex_word_count,
                "Word Count": total_words,
                "Syllables Per Word": syllables_per_word,
                "Personal Pronouns": personal_pronouns,
                "Avg Word Length": avg_word_length
            })
        except Exception as e:
            print(f"Error processing {file_name}: {e}")

# Write results to CSV
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = [
        "File Name", "Positive Score", "Negative Score", "Polarity Score", "Subjectivity Score",
        "Avg Sentence Length", "Percentage of Complex Words", "Fog Index", "Avg Words Per Sentence",
        "Complex Word Count", "Word Count", "Syllables Per Word", "Personal Pronouns", "Avg Word Length"
    ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(results)

print(f"Results have been written to {output_csv}")
```
